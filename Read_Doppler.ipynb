```python
import numpy as np
import math
import matplotlib.pyplot as plt

dt = 10**-6
numsteps = 20000
vx = []
vy = []
vz = []
times = []

f = open("doppler_vel.dat", "rb")

count = 0
for i in range(numsteps*3):
    data_byte = f.read(13) 
    if float(data_byte) >= 0:  #if value is non-negative, only read 12 characters
        f.seek(-13,1)
        data_byte = f.read(12)
        this_vel = float(data_byte)
    else:  #if value is negative, read 13 characters
         this_vel = float(data_byte)
    f.read(1)
    if count%3==0:
        times.append((count/3)*dt)
        vx.append(this_vel)
    elif count%3==1:
        vy.append(this_vel)
    else:
        vz.append(this_vel)
    count+=1
```

```python
plt.figure()
plt.plot(1.0e3 * np.array(times), vx)
plt.xlabel(r'$t/\rm{ms}$')
plt.ylabel(r'$v_x/(\rm{m}/\rm{s})$')
print("Final velcoity in x-direction: " + str(vx[-1]))
```

```python
plt.figure()
plt.plot(1.0e3 * np.array(times), vy)
plt.xlabel(r'$t/\rm{ms}$')
plt.ylabel(r'$v_y/(\rm{m}/\rm{s})$')
print("Final velcoity in y-direction: " + str(vy[-1]))
```

```python
plt.figure()
plt.plot(1.0e3 * np.array(times), vz)
plt.xlabel(r'$t/\rm{ms}$')
plt.ylabel(r'$v_z/(\rm{m}/\rm{s})$')
print("Final velcoity in z-direction: " + str(vz[-1]))
```
